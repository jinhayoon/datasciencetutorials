```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X_full, y_full = datasets.load_diabetes(return_X_y=True)
print(X_full.shape)
print(X_full[0])
print(type(X_full))
```

```python
# Select the 3rd column (index 2) from the full dataset for single-feature regression
# and reshape it to a 2D array (n_samples, 1).
X = X_full[:, np.newaxis, 2]
# Use the full target variable.
y = y_full
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('Diabetes Progression Against BMI')
plt.show()
```
